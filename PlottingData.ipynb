### Problem 10 (4 points)

The variables `q10` is initialized with 5 steps of the Collatz sequence starting from 15.
Collatz sequences work with the following rule.

```
if the current value x is odd:
    the next value will be 3x+1
else:
    the next value will be x/2
```

Extend `q10` to ten entries following this rule, then plot them as a line chart.
Use the values in `q10` as the y values.
You can use the default values for x in the `plt.plot` call.

````python
# DO NOT CHANGE

q10 = [15, 46, 23, 70, 35]
````

````python


while len(q10) < 10:
    current_value = q10[-1]
    if current_value % 2 == 0:
        next_value = current_value // 2
    else:
        next_value = 3 * current_value + 1
    q10.append(next_value)

plt.figure(figsize=(8, 6))
plt.plot(q10, marker='o', linestyle='-', color='purple')
plt.title('Extended Collatz Sequence')
plt.xlabel('Step')
plt.ylabel('Value')
plt.grid(True)
plt.show()
````

### Problem 17 (5 points)

Plot the factorial function from n=1 to 10 using log scale for better legibility.

Recall that the factorial of $n$, denoted $n!$, is calculated as the product $1 \times 2 \times \ldots \times n$.
The plots that you plot should be $(1, 1!), (2, 2!), (3, 3!), \ldots, (10, 10!)$.
Use [`plt.yscale`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.yscale.html) to change your y axis to use log scale.

````python
# YOUR CHANGES HERE


import math


n_values = list(range(1, 11))
factorial_values = [math.factorial(n) for n in n_values]


plt.figure(figsize=(10, 6))
plt.plot(n_values, factorial_values, marker='o', linestyle='-', color='blue')
plt.yscale('log')
plt.title('Factorial Function with Log Scale')
plt.xlabel('n')
plt.ylabel('n! (log scale)')
plt.grid(True)
plt.show()
````

````python



f20_file_path = 'f20.tsv'
try:
    f20_data = pd.read_csv(f20_file_path, sep='\t')

  
    unique_colors = f20_data['color'].unique()

 
    plt.figure(figsize=(10, 6))
    for color in unique_colors:
        subset = f20_data[f20_data['color'] == color]
        plt.scatter(subset['x'], subset['y'], label=color, alpha=0.7)

    plt.title('Scatter Plot of x and y with Colors from f20.tsv')
    plt.xlabel('x values')
    plt.ylabel('y values')
    plt.legend()
    plt.grid(True)
    plt.show()

except FileNotFoundError:
    print("The file 'f20.tsv' was not found. Please upload the file.")
````
